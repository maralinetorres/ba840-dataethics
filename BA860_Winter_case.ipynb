```python
winter_data = pd.read_csv('Winters Digital Attribution.csv')
winter_data.head()
```

```python
winter_data['Position Name'].unique()
```

```python
last_touchpts = winter_data.loc[winter_data['Position Name'] == 'CONVERTER', ]
agg = last_touchpts.groupby('Group Name')[['Order Id']].count().reset_index()
agg['Percentage'] = agg['Order Id'] / agg['Order Id'].sum()
agg.sort_values(by = 'Percentage', ascending=False)
```

```python
first_touchpts = winter_data.loc[winter_data['Position Name'] == 'ORIGINATOR', ]
agg1 = first_touchpts.groupby('Group Name')[['Order Id']].count().reset_index()
agg1['Percentage'] = agg1['Order Id'] / agg1['Order Id'].sum()
agg1.sort_values(by = 'Percentage', ascending=False)
```

```python
first_touchpts['Group Name'].value_counts()
```

Both attributtion models put the CPM channel with the hightest attribution. However, Last touch points doubles the channel attribution. 

------------------

Focusing in new customers



```python
last_touchpts1 = winter_data.loc[(winter_data['Position Name'] == 'CONVERTER') & (winter_data['New Customer'] == 'Y'), ]
agg2 = last_touchpts1.groupby('Group Name')[['Order Id']].count().reset_index()
agg2['Percentage'] = agg2['Order Id'] / agg2['Order Id'].sum()
agg2.sort_values(by = 'Percentage', ascending=False)
```

```python
first_touchpts1 = winter_data.loc[(winter_data['Position Name'] == 'ORIGINATOR') & (winter_data['New Customer'] == 'Y'), ]
agg3 = first_touchpts1.groupby('Group Name')[['Order Id']].count().reset_index()
agg3['Percentage'] = agg3['Order Id'] / agg3['Order Id'].sum()
agg3.sort_values(by = 'Percentage', ascending=False)
```

```python
## Merge to see if we can compare
compare_last = pd.merge(agg, agg2, how ='outer', on='Group Name', suffixes=['_all','_new'])
compare_last[['Group Name','Percentage_all','Percentage_new']]
```

```python
compare_first = pd.merge(agg1, agg3, how ='outer', on='Group Name', suffixes=['_all','_new'])
compare_first[['Group Name','Percentage_all','Percentage_new']]
```

```python
### OLD CUSTOMERS

last_touchpts2 = winter_data.loc[(winter_data['Position Name'] == 'CONVERTER') & (winter_data['New Customer'] == 'N'), ]
agg2 = last_touchpts2.groupby('Group Name')[['Order Id']].count().reset_index()
agg2['Percentage'] = agg2['Order Id'] / agg2['Order Id'].sum()
agg2.sort_values(by = 'Percentage', ascending=False)
```

```python
first_touchpts2 = winter_data.loc[(winter_data['Position Name'] == 'ORIGINATOR') & (winter_data['New Customer'] == 'N'), ]
agg3 = first_touchpts2.groupby('Group Name')[['Order Id']].count().reset_index()
agg3['Percentage'] = agg3['Order Id'] / agg3['Order Id'].sum()
agg3.sort_values(by = 'Percentage', ascending=False)
```

```python
## Merge to see if we can compare
compare_last = pd.merge(agg, agg2, how ='inner', on='Group Name', suffixes=['_all','_old'])
compare_last[['Group Name','Percentage_all','Percentage_old']]
```

```python
compare_first = pd.merge(agg1, agg3, how ='inner', on='Group Name', suffixes=['_all','_old'])
compare_first[['Group Name','Percentage_all','Percentage_old']]
```

#### TIME - DECAY


```python
winter_data['Non-normalized credits'] = 2**-winter_data['Time to Convert (Days)']/7
winter_data[['Group Name','Non-normalized credits','Time to Convert (Days)']]
```

```python
grouped = winter_data[['Order Id','Non-normalized credits']].groupby('Order Id')
# grouped['Normalize credits'] = grouped['Non-normalized credits'].transform(lambda x: x/sum(x))
# grouped[['Group Name','Normalize credits','Time to Convert (Days)']]
winter_data['t_decay_credit_normed'] = grouped.transform(lambda x: x/sum(x))
```

```python
winter_data.head()
```

```python
num_order = len(winter_data['Order Id'].unique())
T_tdecay = pd.DataFrame(winter_data[['Group Name', 't_decay_credit_normed']].groupby('Group Name').sum()/num_order).reset_index()
```
